```python
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
def sigmoid_derivative(x):
    return x*(1-x)
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
    ])
```

```python
y=np.array([[0],[1],[1],[0]])
```

```python
np.random.seed(42)
```

```python
W1 = np.random.randn(2, 3) # 3X2
W1
```

```python
W2 = np.random.randn(3, 1) # 3X1 
W2
```

```python
lr = 0.1
```

# 3 layers
- > input
- > hidden
- > output

```python
epochs = 10000
for epoch in range(epochs):

    # feed forward pass
    hidden_input = np.dot(X, W1) #dot product of X and W1([w1,w2])
    hidden_output = sigmoid(hidden_input) # f(x)

    final_input = np.dot(hidden_output, W2) #
    final_output = sigmoid(final_input)

    # loss calculation
    error = y - final_output # -> error matrix ([[],[],[],[]])

    # backpropogation pass
    d_output = error * sigmoid_derivative(final_output) 
    d_hidden = d_output.dot(W2.T) * sigmoid_derivative(hidden_output) # 

    # Update weights
    W2 = W2 + hidden_output.T.dot(d_output) * lr
    W1 = W1 + X.T.dot(d_hidden) * lr
```
